#Clase 4: Series Temporales

## Componentes

Una serie temporal es una secuencia cronológica ordenada de observaciones de una variable particular.

En las series temporales se puede identificar los siguientes componentes:
- tendencia: comportamiento a largo plazo. Nos muestra la dirección en la que se mueven/cambian los valores. Puede ser un valor creciente o decreciente.
- estacionalidad: variación regular que se repite durante un período específico.
- aleatoriedad: variación errática o residual.

A continuación, veremos un ejemplo con datos reales. Para ello, en primer lugar importamos las librerías necesarias.

```python
import numpy as np
import pandas as pd
import datetime

from matplotlib import pyplot as plt
import seaborn as sns

import requests
import json
```

Analizaremos los datos proveídos por el **INA**: la altura hidrométrica registrada en la estación 86, La Plata.

```python
# Hacemos un get request
timeStart = "2020-01-01"
timeEnd = "2024-03-30"

url_api =(f"https://alerta.ina.gob.ar/pub/datos/datos&timeStart={timeStart}&timeEnd={timeEnd}&seriesId=86&siteCode=86&varId=2&format=json")

response = requests.get(url_api)
datos = response.json()

df = pd.DataFrame.from_dict(datos["data"])
df.head()
```

Una vez levantados los datos, los limpiamos:

```python
# Nos quedamos con las columnas que nos interesan y las renombramos
df = df[['timestart', 'valor']].rename(columns={'timestart':'fecha', 'valor': 'altura'})
df.head()
```

```python
df.info()
```

```python
# Cambiamos el formato de los datos de "fecha"
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
df.info()
```

```python
# Creamos nuevas columnas que pueden ser de utilidad
df['año'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
df['semana'] = df['fecha'].dt.week
df['dia'] = df['fecha'].dt.day
```

```python
df.head()
```

### Análisis de la serie temporal

En primer lugar, graficamos la serie temporal para conocer las obervaciones.

**¿Nota alguna tendencia o estacionalidad?**

```python
plt.figure(figsize=(18, 6))

sns.lineplot(data = df, x='fecha', y='altura'
).set(title='Serie temporal: Altura hidrométrica, estación La Plata', ylabel='Altura (m)', xlabel='Fechas')
```

Una forma sencilla de analiza si existe estacionalidad es con **boxplots**:

```python
sns.boxplot(data=df, x='mes', y='altura', showfliers=False)
```

Otra forma es realizando un **remuestreo**, que puede llevarse a cabo con la conocida `.groupby()` o `.resample()`:

```python
# Agrupamos por mes de cada año
df_mean = df.groupby(['año', 'mes'])[['altura']].agg('mean')
df_mean.head()
```

```python
# Primero, tenemos que colocar la fecha como índice antes de remuestrar
df_mean = df[['fecha', 'altura']].set_index("fecha")
df_mean = df_mean.resample("M").mean()
df_mean.head()
```

```python
# Graficamos
plt.figure(figsize=(18, 6))

sns.lineplot(data = df, x='fecha', y='altura', label= 'Altura diaria'
).set(title='Serie temporal: Altura hidrométrica, estación La Plata', ylabel='Altura (m)', xlabel='Fechas')

sns.lineplot(df_mean, x='fecha', y='altura', marker='o', label='Remuestreo: Altura media mensual')
```

Otra opción, es calular la **media móvil**.

Cuando la variabilidad es lenta y gradual, una forma sencilla de visualizar estas tendencias es con medias móviles en diferentes escalas de tiempo. Una media móvil tiende a suavizar una serie de tiempo promediando variaciones en una ventana de tiempo igual a su longitud.

Usemos el método `.rolling()` para calcular la media móvil de 7 y 30 días de nuestros datos diarios. Usamos el argumento `center=True` para etiquetar cada ventana en su punto medio, resultando:

```python
# Ejemplo: no centrado
df['altura'].rolling(3).mean().head()
```

```python
# Ejemplo: centrado
df['altura'].rolling(3, center=True).mean().head()
```

```python
# Media móvil con ventana de 7 días
df_7d = df.set_index('fecha').copy()
df_7d = df_7d['altura'].rolling(7, center=True).mean()

# Media semanal
df_mean = df.set_index("fecha")
df_mean = df_mean.resample("W").mean().reset_index()


# Gráficos
plt.figure(figsize=(18, 6))

sns.lineplot(data = df, x='fecha', y='altura', label='Altura diaria'
).set(title='Serie temporal', ylabel='Altura hidrométrica', xlabel='Fechas')

sns.lineplot(data = df_mean, x='fecha', y='altura', marker='o', label='Remuestreo con la altura media semanal')

sns.lineplot(df_7d, label='Rolling 7 dias')

#plt.xlim([datetime.date(2020, 6, 1), datetime.date(2020, 12, 1)])
```

```python
# Media móvil con ventana de 30 días
df_30d = df.set_index('fecha').copy()
df_30d = df_30d['altura'].rolling(30, center=True).mean()

# Media semanal
df_mean = df.set_index("fecha")
df_mean = df_mean.resample("M").mean().reset_index()
df_mean


# Gráficos
plt.figure(figsize=(18, 6))

sns.lineplot(data = df, x='fecha', y='altura', label='Altura diaria'
).set(title='Serie temporal', ylabel='Altura hidrométrica', xlabel='Fechas')

sns.lineplot(data = df_mean, x='fecha', y='altura', marker='o', label='Remuestreo con la altura media mensual')

sns.lineplot(df_30d, label='Rolling 30 dias')

#plt.xlim([datetime.date(2020, 12, 1), datetime.date(2021, 12, 1)])
```

Finalmente, podríamos realizar un **gráfico de autocorrelación**. Este permite verificar la aleatoriedad en un conjunto de datos, midiendo si se correlacionan los valores actuales contra los pasados:

$$\rho _{XX}(t_{1},t_{2}) = \tfrac{E\left[ ( X_{t1}-\mu_{t1} )\cdot \left( X_{t2}-\mu _{t2}\right) \right]}{\sigma_{t1}\sigma_{t2}}
$$

Es decir, se calcula la correlación entre un valor y la versión desplazada en el tiempo. Ese desplazamiento temporal, se conoce como desfase o lag.

```python
pd.plotting.autocorrelation_plot(df["altura"])
#plt.ylim(-0.25,0.25)

# ¿Cuál cree que es el patrón de estacionalidad
#plt.axvline( , c="r")
```
