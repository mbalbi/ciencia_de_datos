```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/Titanic.csv'
datos = pd.read_csv(url, sep = ',')
```

```python
datos
```

```python
# Siguiendo los pasos
# 1) Importar librerias
# import seaborn as sns
# import numpy as np
# import matplotlib.pyplot as plt

# 2) Crear objeto de figura
plt.figure()

# 3) Dibujar
# BARPLOT
sns.barplot(data = datos,
           x = 'Pclass',
           y = 'Survived',
            ci=None)

#4) Mostrar
plt.show()
```

```python
plt.figure()
sns.lineplot(data = datos,
             x = 'Pclass',
             y = 'Fare',
             errorbar=('ci', 90)
             )
plt.show()
```

```python
# EJEMPLO 3) SCATTERPLOT
plt.figure()
sns.scatterplot(data = datos,
                x = 'Age',
                y = 'Fare')
plt.show()
```

```python
# EJEMPLO 4) HISTOGRAMA
plt.figure()
sns.histplot(data = datos,
                x = 'Age')
plt.show()
```

```python
plt.figure()
sns.histplot(data = datos,
             x = 'Age',
             alpha = 0.2,
             edgecolor = '#B46DE5',
             stat = 'probability',
             bins = 20)
plt.show()
```

```python
# PARÁMETRO HUE
plt.figure()
sns.histplot(data = datos,
             x = 'Age',
             alpha = 0.6,
             edgecolor = '#6B7498',
             stat = 'probability',
             bins = 20,
             hue = 'Survived')
```
