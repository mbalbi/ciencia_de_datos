## Ejercicio 2: Modelo digital del terreno

Ahora que sabemos como funcionan bien los numpy arrays y sabemos hacer operaciones con ellos, veamos un caso de aplicación de todo esto.

Para evaluar el comportamiento de una cuenca necesitamos conocer su topografía. Ahora, no vamos a tener medida la elevación de todos los puntos, por lo que deberemos generar un Modelo Digital del Terreno en base de algunos puntos medidos, e infiriendo el resto de alguna forma lógica. Un método habitual para hacer eso es, por ejemplo, el de distancia inversa ponderada (IDW). En este método, la elevación de un punto $i$ cualquiera con coordenadas $x_{i}$, $y_{i}$ se puede calcular de la siguiente forma:

$$z_{i} = \frac{\sum_{j=1}^{n} \frac{1}{d_{ij}^{p}}z_{j}}{\sum_{j=1}^{n} \frac{1}{d_{ij}^{p}}}$$

donde $p$ es un parámetro a definir (por ejemplo, $p=3$) que define cuán más importante es la cercanía al punto a interpolar, y $d_{i,j}$ es la distancia entre el punto $i$ cuya altura se quiere definir y cada punto $j$ cuya altura se conoce:

$$d_{i,j} = \sqrt{\left(x_{i}-x_{j}\right)^{2}+\left(y_{i}-y_{j}\right)^{2}} = \bar{X}_{i}\cdot \bar{X}_{j} = \bar{X}_{j}^{T}\bar{X}_{i}$$

Entonces, un punto clave del proceso de construir nuestro modelo digital es poder construir la matriz $D$ que tiene en cada fila la distancia del punto $i$ con todos los puntos $j$.

El objetivo de este ejercicio es que vean como a medida que se requiere un mayor número de operaciones, la eficiencia del algoritmo empieza a pesar más y convertir a un problema de viable a inviable (o viceversa).

Para ello calcularemos la aceleración rms móvil en ventanas de 5 segundos de duración. Esto lo podemos hacer con pandas. No es tan fácil como para pedirles a ustedes que lo hagan ahora, pero sí les explicaremos qué se hace:

1. Se toma el cuadrado de las aceleraciones: `df_vibraciones['a(t) [m/s^2']**2`
2. A ello, se le toma en forma móvil la cantidad de datos que se corresponden a 5 segundos: `idxvent = int(5//df_vibraciones['t [s]'].values[1]-1)` es el numero de datos, y al resultado de 1. le aplico `.rolling(idxvent)`
3. Al resultado de 2. le calculo la media y le tomo la raíz cuadrada `.mean**0.5`

Un choclo bastante feo, pero lo importante es que es una medida del nivel de aceleraciones en un tiempo de 5 segundos.

```python
# Calcular el rms movil
vent = 5 # Longitud de ventana en segundos
idxvent = int(vent // df_vibraciones['t [s]'].values[1] - 1)
df_vibraciones['Rolling_rms'] = (df_vibraciones['a(t) [m/s^2]']**2).rolling(idxvent).mean() **0.5
df_vibraciones.head()
```
