Verificamos que solo nos hayan quedado un nombre por ciudad.

```python
df_nombres_corregidos['Superfluidificante']=df_nombres_corregidos['Superfluidificante'].replace({
 0:'No',
 1:'Si'
})
```

Chequeamos que efectivamente esté bien aplicado el cambio utilizando el metodo .info().

```python
df_nombres_corregidos['Superfluidificante'].unique()
```

## 4.3Aplicación de funciones

Las operaciones que acabamos de ver son relativamente simples. Pero en otros casos podríamos tener que aplicar una o más operaciones simultaneamente, para lo cual vale la pena trabajar con funciones.

Lo ideal sería definir la función por separado y después aplicarla en nuestro dataframe. En este caso vamos a utilizar la ley de Powers, que vincula relación agua-cemento con resistencia.

$ f_c = A \cdot \, e^{-B \ \cdot \left(\tfrac{w}{c}\right)}$

donde 120 y 2.77 son valores razonables para A y B respectivamente. En este caso, haremos que A y B se calculen cada vez siguiendo una distribución normal con una relación entre media y desvío de 10%.


```python
def powers(ac):
    A=np.random.normal(120,120/10)
    B=np.random.normal(2.77,2.77/10)
    return A*np.exp(-B*ac)
```

Ahora aplicaremos esta función sobre todas las filas de nuestro dataframe para obtener la resistencia del hormigón segun Powers. La guardamos en una columna llamada "fc_mod_Mpa" ya que es la resistencia del hormigón modelada y está en MPa.

```python
df_base['fc_mod_MPa']=df_base.apply(lambda x:powers(x['agua/cemento']),axis=1)
```

Verificamos que se haya creado sin problemas nuestra nueva variable y que tome valores razonables.

```python
df_base
```

# 5.Definición de los tipos de datos

En esta sección veremos como convertir las distintas columnas que tenemos a los tipos de datos que les corresponden. Por ejemplo, la columna "Fecha" hasta este punto era considerada un texto. Sin embargo, pandas tiene la capacidad de reconocerla con su propio formato lo cual puede ser muy útil posteriormente. A continuación le daremos a columna el formato que le corresponde (en la mayoría de los casos, ya lo tienen por default).


```python
df_base.columns
```

```python
df_base['Origen']=df_base['Origen'].astype('category')
df_base['Cemento_m3']=df_base['Cemento_m3'].astype('float')
df_base['Agua_m3']=df_base['Agua_m3'].astype('float')
df_base['Arena_m3']=df_base['Arena_m3'].astype('float')
df_base['Piedra_m3']=df_base['Piedra_m3'].astype('float')
df_base['Superfluidificante']=df_base['Superfluidificante'].astype('category')
df_base['Asentamiento_cm']=df_base['Asentamiento_cm'].astype('float')
df_base['Resistencia_MPa']=df_base['Resistencia_MPa'].astype('float')

df_base['Fecha']=pd.to_datetime(df_base['Fecha'],format='%d/%m/%Y')

df_base['agua/cemento']=df_base['agua/cemento'].astype('float')
df_base['Prop_AG']=df_base['Prop_AG'].astype('float')
df_base['fc_mod_MPa']=df_base['fc_mod_MPa'].astype('float')
df_base.info()
```

Haber convertido nuestras fechas al formato correcto nos permite facilmente crear una nueva columna que tenga solo el año del hormigón por ejemplo.

```python
df_base['Year']=df_base['Fecha'].dt.year
df_base
```

# 6.Utilización de filtros

Ahora que ya sabemos cómo limpiar y organizar nuestros datos, es momento de aprender a trabajar solo con la información que nos interesa. Para esto usamos los filtros. Usar filtros nos ayuda a analizar de manera más sencilla y ordenada nuestro conjunto de datos. Veamos cómo hacerlo con algunos ejemplos.

## 6.1Filtrado según una columna específica

En este caso vamos a crear un dataframe que contenga unicamente los hormigones elaborados en la planta de Bahía Blanca.

```python
df_bahia=df_base.copy()
df_bahia=df_bahia[df_bahia['Origen']=='Bahia Blanca']
```

Podemos verificar que la única ciudad presente en el dataframe es Bahía Blanca. Sin embargo, como definimos la categoría origen como Categoría anteriormente, nos va a mostrar primero los valores presentes en el dataframe para esa columna y después los valores posibles para esa categoría.

```python
df_bahia['Resistencia_MPa'].mean()
df_base[df_base['Origen']=='Mercedes']['Resistencia_MPa'].min()
```

## 6.2Filtrando con más de una condición

En esta sección vamos a ver como establecer varias condiciones a nuestro filtro. Para ello, vamos a hacerlo en dos pasos para poder hacerlo en forma mas prolija. Vamos a filtrar aquellos hormigones elaborados:
- sin superfluidificante
- con resistencia mayor a 25MPa
- con origen en Bahía Blanca

```python
df_filtromultiple=df_base.copy()
filtro=((df_base['Superfluidificante']=='No') & (df_base['Resistencia_MPa']>25)&(df_base['Origen']=='Bahia Blanca'))
df_filtromultiple=df_filtromultiple[filtro]
df_filtromultiple
```

## 6.3Filtro utilizando más de un valor

Veamos como filtrar utilizando más de un valor por columna de distintas maneras.

```python
df_variosvalores=df_base.copy()
listaciudades=['Olavarria','Rio Cuarto']
df_variosvalores=df_variosvalores[df_variosvalores['Origen'].isin(listaciudades)]
df_variosvalores['Origen'].unique()

filtrociudades = ((df_base['Origen'] == 'Olavarria') | (df_base['Origen'] == 'Rio Cuarto'))
df_variosvalores=df_base[filtrociudades]
df_variosvalores
```

Nuevamente, verificamos que nuestro dataframe solo contenga filas según nuestro filtro.

## 6.4Filtro según fecha

Por último, vamos a ver como filtrar en nuestro dataset según fechas (a esta altura ya tenemos nuestra columna fecha correctamente tipificada). En este caso vamos a filtrar valores del primer cuatrimestre del 2024 (1/1 al 30/4)

```python
df_1c2024=df_base.copy()
inicio = "2024-01-01"
fin = "2024-04-30"

df_1c2024=df_1c2024[(df_1c2024['Fecha']>=inicio) & (df_1c2024['Fecha']<=fin)]
df_1c2024
```
